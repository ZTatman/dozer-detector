# Historical Model and Demo Assets

This notebook inspects the recovered 2021 TensorFlow.js browser demo assets. It does not claim to reproduce the original training run. The goal is to document what survived and provide a basis for a modern demo cleanup.

```python
from pathlib import Path
import json

ROOT = Path.cwd()
DEMO = ROOT / 'demo' / 'tfjs'
MODEL = DEMO / 'zach_model' / 'model.json'

print('Demo exists:', DEMO.exists())
print('TFJS model exists:', MODEL.exists())
```

## Inspect Model Metadata

```python
model = json.loads(MODEL.read_text())
print('format:', model.get('format'))
print('generatedBy:', model.get('generatedBy'))
print('convertedBy:', model.get('convertedBy'))
print('weightsManifest entries:', len(model.get('weightsManifest', [])))

weight_files = sorted((DEMO / 'zach_model').glob('*.bin'))
for path in weight_files:
    print(path.name, path.stat().st_size)
```
